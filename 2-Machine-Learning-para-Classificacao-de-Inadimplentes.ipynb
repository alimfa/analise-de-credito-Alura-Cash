```python
#identifica linhas que a renda é igual a zero para entender se estes valores estão corretos

df[df.person_income == 0]
```

```python
#Corrige linhas com renda igual a zero, porém com 'loan_percent_income' diferente de zero
mask = df[df.person_income == 0][df.loan_percent_income != 0][df.loan_amnt != 0].index
df.loc[mask, 'person_income'] = df.loan_amnt/df.loan_percent_income
```

```python
df[df.loan_amnt == 0]
```

```python
#Verifica valores errados de 'loan_percent_income'

df['validacao'] = round(df.loan_amnt/df.person_income, 2)
df['diferenca'] = round(df.validacao - df.loan_percent_income, 2)
df[df['diferenca'] != 0.0]
```

```python
#Corrige valores de 'loan_percent_income'

df['loan_percent_income'] =  round(df.loan_amnt/df.person_income, 2)
df['loan_percent_income']= df['loan_percent_income'].replace(np.inf, 0.00)

df = df.drop(['validacao','diferenca'], axis=1)
```

## Determinação da Linha de Base

```python
#Calcula acurácia usando Dummy Classifier

SEED = 123
np.random.seed(SEED)

modelo = DummyClassifier()
results = cross_validate(modelo, x, y, cv = 10, return_train_score=False)
media = results['test_score'].mean()
desvio_padrao = results['test_score'].std()
print("Acurácia com dummy stratified, 10 = [%.2f, %.2f]" % ((media - 2 * desvio_padrao)*100, (media + 2 * desvio_padrao) * 100))
```

## Validação de Modelos

Agora o intuíto é testar 4 diferentes modelos de Machine Learning utilizando Cross Validation com aleatoriedade para cada um deles com o objetivo de identificar qual apresentará melhor range possível no valor de suas métricas. 

As métricas analisadas serão: Acurácia, Precisão, Sensibilidade (Recall) e F1. 

```python
#Cria função para imprimir as métricas


def imprime_resultados(results):
    media1 = results['test_accuracy'].mean()
    media2 = results['test_precision'].mean()
    media3 = results['test_recall'].mean()
    media4 = results['test_f1'].mean()
    desvio_padrao1 = results['test_accuracy'].std()
    desvio_padrao2 = results['test_precision'].std()
    desvio_padrao3 = results['test_recall'].std()
    desvio_padrao4 = results['test_f1'].std()
    print("Acurácia ficará entre = [%.2f, %.2f]" % ((media1 - 2 * desvio_padrao1) * 100, (media1 + 2 * desvio_padrao1) * 100))
    print("Precisão ficará entre = [%.2f, %.2f]" % ((media2 - 2 * desvio_padrao2) * 100, (media2 + 2 * desvio_padrao2) * 100))
    print("Sensibilidade ficará entre = [%.2f, %.2f]" % ((media3 - 2 * desvio_padrao3) * 100, (media3 + 2 * desvio_padrao3) * 100))
    print("F1 ficará entre = [%.2f, %.2f]" % ((media4 - 2 * desvio_padrao4) * 100, (media4 + 2 * desvio_padrao4) * 100))
```

```python
#Define Seed e aleatoriedade no Cross Validation usando Kfold

cv = KFold(n_splits = 10, shuffle = True)
```

### Linear SVC

```python
#Valida o modelo LinearSVC

modelo =  LinearSVC(max_iter=1000, random_state=SEED)
results = cross_validate(modelo, x, y, cv = cv, return_train_score=False, scoring=('accuracy', 'precision', 'recall', 'f1'))
imprime_resultados(results)
```

### K-Neighbors Classifier

```python
#Valida o modelo K-Neighbors Classifier

modelo =  KNeighborsClassifier(n_neighbors = 9)
results = cross_validate(modelo, x, y, cv = cv, return_train_score=False, scoring=('accuracy', 'precision', 'recall', 'f1'))
imprime_resultados(results)
```

### Decision Tree

```python
#Valida o modelo Decision Tree Classifier
modelo = DecisionTreeClassifier(max_depth= 10, random_state=SEED)
results = cross_validate(modelo, x, y, cv = cv, return_train_score=False, scoring=('accuracy', 'precision', 'recall', 'f1'))
imprime_resultados(results)
```

### Random Forest

```python
#Valida o modelo Random Forest Classifier
modelo = RandomForestClassifier(max_depth= 10, random_state=SEED)
results = cross_validate(modelo, x, y, cv = cv, return_train_score=False, scoring=('accuracy', 'precision', 'recall', 'f1'))
imprime_resultados(results)  
```

### Considerações

Entende-se que fornecer crédito a pessoas potencialmente inadimplentes trará grandes prejuízos à empres, por isso, é interessante que o número de falsos negativos seja o menor possível. 
Além disso, é altamente desejável que o modelo consiga identificar aqueles clientes que serão inadimplentes com maior exatatidão, mesmo que para isso, o modelo classifique erroneamente alguns clientes que não seriam inadiplentes (falsos positivos).

Portanto, o modelo que apresentou menor número de falsos negativos foi o K-Neighbors Classifier, tendo apresentado também todas as quatro métricas (acurácia, precisão, recall e f1) mais equeilibradas entre si, e acima de 0,83.  

Desta forma, o modelo que utiliza o **K-Neighbors Classifier**  foi o escolhido para ser otimizado, com foco em melhorar principalmente a sua **acurácia** e **sensibilidade**, a fim de reduzir ao máximo os falsos negativos.

## Otimização de Hiperparametros


```python
#Cria grid de hiperparametros

grid = { 'n_neighbors': [3, 5, 9, 11, 15, 23],
           'weights' : ['uniform', 'distance']}
```

```python
#Busca melhor configuração do modelo KNN

busca_melhor = GridSearchCV(KNeighborsClassifier(),
                    grid,
                    cv = KFold(n_splits = 10, shuffle = True))

busca_melhor.fit(x, y)
busca_melhor.best_params_
```

```python
#Imprime os resultados possíveis dentro do range do Cross Validation
modelo_otimizado =  KNeighborsClassifier(**busca_melhor.best_params_)
results = cross_validate(modelo_otimizado, x, y, cv = cv, return_train_score=False, scoring=('accuracy', 'precision', 'recall', 'f1'))
imprime_resultados(results)
```

## Analisando Matriz Confusão e Curva ROC do modelo otimizado

```python
#Separa x e y para treino e para teste, sendo 25% da amostra para teste
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state = SEED)

#Treina o modelo KNeighbors Classifier
knn = KNeighborsClassifier(**busca_melhor.best_params_)
knn.fit(x_train, y_train)
y_pred = knn.predict(x_test)

#Imprime as métricas encontradas para KNeighbors Classifier
print("Acurácia:",metrics.accuracy_score(y_test, y_pred))
print("Precisão:",metrics.precision_score(y_test, y_pred))
print("Recall:",metrics.recall_score(y_test, y_pred)) 
print("F1:",metrics.f1_score(y_test, y_pred))
```

```python
#Cria a Matriz Confusão 

cm = confusion_matrix(y_test, y_pred)
```

```python
#Plota a Matriz Confusão 


disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()

plt.savefig('.../Graficos/matriz-confusao.png', format='png')
plt.savefig('.../Graficos/matriz-confusao.jpeg', format='jpeg')
```

```python
#Plota Curva ROC
y_pred_proba = knn.predict_proba(x_test)[::, 1]
fpr1, tpr1, _= metrics.roc_curve(y_test, y_pred_proba)
auc = metrics.roc_auc_score(y_test, y_pred_proba)
plt.rcParams['figure.figsize'] = (12., 8.)
plt.plot(fpr1,tpr1,color = 'orange', label="ROC,   AUC="+str(auc))
plt.plot([0, 1], [0, 1], color='blue', lw=2, linestyle='--', label ='Guessing')
plt.xlabel('False positive rate (fpr)')
plt.ylabel('True positive rate (tpr)')
plt.title('RECEIVER OPERATING CHARACTERISTIC (ROC) CURVE')
plt.legend(loc=4)
plt.show()

plt.savefig('.../Graficos/curvaROC.png', format='png')
plt.savefig('.../Graficos/curvaROC.jgep', format='jpeg')
```
